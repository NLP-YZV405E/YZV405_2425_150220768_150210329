```python
def csv_to_tsv(input_csv, output_tsv):
    df = pd.read_csv(input_csv)

    with open(output_tsv, 'w', encoding='utf-8') as fout:
        for _, row in df.iterrows():
            switch = False
            tokens = ast.literal_eval(row['tokenized_sentence'])
            indices = ast.literal_eval(row['indices'])
            
            if indices == [-1]:
                tags = ['O'] * len(tokens)
            else:
                tags = ['O'] * len(tokens)
                for i in indices:
                    if not switch:
                        tags[i] = 'B-IDIOM'
                        switch = True
                    else:
                        tags[i] = 'I-IDIOM'
            if tokens[-1] != '.':
                tokens.append('.')
                tags.append('O')
            for token, tag in zip(tokens, tags):
                
                fout.write(f"{token}\t{tag}\n")
```

```python
csv_to_tsv(r"./train.csv", r"./train.tsv")
csv_to_tsv(r"./eval.csv", r"./dev.tsv")
```
